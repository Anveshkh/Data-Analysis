```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# The plot function is used to plot points in a diagram
# By default the plot function draws a line from point to point.

X_points = np.array([1,2,3,4,5,6])
Y_points = np.array([1,4,9,16,25,36])
plt.plot(X_points, Y_points);
```

```python
# If we only want to show the points or markers
plt.plot(X_points, Y_points, 'o')
```

```python
x1 = np.array([1,2,6,8])
x2 = np.array([3,8,1,10])
plt.plot(x1, x2)
```

```python
# If we do not specify the points on the x-axis, they will get the default values from 0 to n

y_points = np.array([3,8,1,10,5,7])
plt.plot(y_points)
```

```python
# only points
plt.plot(y_points, 'o')
```

### Markers

```python
y_points = np.array([3,8,1,10])
plt.plot(y_points, marker='o')
plt.show()
```

```python
plt.plot(y_points, marker='*')
plt.show()
```

### Format Strings fmt
You can use the shortcut string notation parameter to specify the marker.
This parameter is called fmt.
marker|line|color

```python
plt.plot(y_points, 'o:r')
```

```python
y_pts = np.array([1,4,9,16,25,36])
plt.plot(y_pts, '*:b')
```

```python
plt.plot(y_pts, '*-g');
```

### Linestyle

You can use the keyword argument linestyle, or shorter ls, to change the style of the plotted line.

```python
y_points = np.array([3,8,1,10])
plt.plot(y_points, '*--y');
```

```python
plt.plot(y_points, linestyle='dotted')
```

```python
plt.plot(y_points, ls='--')
```

```python
plt.plot(y_points, color='r')
```

### MULTIPLE LINES

You can plot as many lines as you like by simply adding more plt.plot() functions.

```python
y1 = np.array([3,8,1,10])
y2 = np.array([6,2,7,11])

plt.plot(y1, '*-b')
plt.plot(y2, '+--r')
plt.show()
```

```python
# Instead of using two plt.plot functions you can use one plot function for both lines

x1 = np.array([0,1,2,3])
y1 = np.array([3,8,1,10])
x2 = np.array([0,1,2,3])
y2 = np.array([6,2,7,11])

plt.plot(x1,y1, '*-r')
plt.plot(x2,y2, '+--b')
plt.show()
```

```python
plt.plot(x1,y1,x2,y2);
```

## Create labels for the plot

```python
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])
plt.plot(x,y, '*-b');
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.title("Sports Watch Data")
plt.show()
```

## Add Grid Lines to the plot
You can use grid() function to add grid lines to the plot

```python
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x,y)
plt.grid()
plt.show()
```

## Display multiple points

With the subplot() function you can draw multiple plots in one figure:

```python
# plot 1
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(1,2,1) # 1 row, 2 columns and this subplot is the 1st one of 2.
plt.plot(x,y)

# plot 2:
x = np.array([0,1,2,3])
y = np.array([10,20,30,40])

plt.subplot(1,2,2)
plt.plot(x,y)

plt.show()
```

\
The `subplot()` function takes three arguments that describes the layout of the figure.
The layout is organized in rows and columns, which are first and second argument of the function.
The third argument represents the index of the current plot

## Draw 2 plots on top of each other

```python
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(2,1,1)
plt.plot(x,y)

x = np.array([0,1,2,3])
y = np.array([10,20,30,40])

plt.subplot(2,1,2)
plt.plot(x,y)

plt.show()
```

## Bar Graph

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3,8,1,10])

plt.bar(x,y)
plt.show()
```

## Horizontal Bars

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3,8,1,10])

plt.subplot(2,1,1)
plt.bar(x,y)
plt.title("Bar Plot")

plt.subplot(2,1,2)
plt.barh(x,y)
plt.title("Horizontal Bar plot")
plt.show()
```

## HISTOGRAMS

A Histogram is a graph showing frequency distributions. It is a graph showing number of observations within each given interval.

```python
# A random data distribution from numpy
x = np.random.normal(170,10,250) # 170 -> mean, 10 -> variance, 250 -> size of array
print(x)
```

```python
plt.hist(x)
plt.show()
```
